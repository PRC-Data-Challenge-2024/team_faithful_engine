```python
import pandas as pd
```

```python
df = pd.read_parquet("/home/wues_ni/Projects/ATOW-Prediction/notebooks/experimental/training.parquet")
preds = pd.read_csv("/home/wues_ni/Projects/ATOW-Prediction/training_predictions.csv")
preds.rename(columns={"tow": "predicted_tow"}, inplace=True)
# preds = pd.read_csv("/home/wues_ni/Projects/ATOW-Prediction/submissions/team_faithful_engine_v1_bfdbc2e9-cc5c-4f10-82eb-9b8e3158e2a3.csv")
```

```python
df = preds.merge(df, on="flight_id")
df["error"] = df["predicted_tow"] - df["tow"]
df["absolute_error"] = df["error"].abs()
```

```python
cols = list(df.columns)
cols.remove("tow")
cols.remove("flight_id")
cols.remove("predicted_tow")
cols.remove("error")
cols.remove("absolute_error")
```

```python
cols = ["flight_id", "tow", "predicted_tow", "error", "absolute_error"] + cols
len(cols)
```

```python
df[cols].sort_values(by="absolute_error", ascending=False).reset_index(drop=True).to_csv("predictions_with_errors.csv")
```
